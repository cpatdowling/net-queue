```python
import os
import numpy as np
import matplotlib.pyplot as plt

datadir = "/home/chase/projects/net-queue/data/simulation/convergence/convg_rate_1/"
output = os.listdir(datadir)

outputs = []
for f in output:
    if f[0:5] == "occup":
        data = np.loadtxt(datadir + f, delimiter=",")
        data = np.mean(data, axis=0)
        outputs.append(data)
        
data = np.mean(np.asarray(outputs), axis=0)

plt.plot(data)
plt.show()
```
