```python
import os
import numpy as np
import matplotlib.pyplot as plt
dataDir = "/home/chase/projects/net-queue/data/"
files = os.listdir(dataDir)
types = ["averageWait", "utilization"]

numRates = 160

def collate_mc_data(dataPath):
    files = os.listdir(dataPath)
    waitArrays = []
    utilArrays = []

    counterWait = 0
    counterUtil = 0
    for fileName in files:
        if fileName[0] == "a":
            arr = np.loadtxt(dataPath + fileName, delimiter=",")
            if arr.shape[0] == numRates:
                waitArrays.append(arr)
        if fileName[0] == "u":
            arr = np.loadtxt(dataPath + fileName, delimiter=",")
            if arr.shape[0] == numRates:
                utilArrays.append(arr)
    
    waitTimes = waitArrays.pop(0)
    utilVals = utilArrays.pop(0)
    for i in waitArrays:
        waitTimes = np.concatenate((waitTimes, i[:,1::]), axis=1)
    for j in utilArrays:
        utilVals = np.concatenate((utilVals, j[:,1::]), axis=1)
        
    return(waitTimes, utilVals)


dataWait2, utilVals2 = collate_mc_data(dataDir + "4bus/")
waitAvg2 = np.concatenate((1.0/dataWait2[:,0:1], np.reshape(np.mean(dataWait2[:,1::], axis=1), (numRates,1))), axis=1)
waitStd2 = np.concatenate((1.0/dataWait2[:,0:1], np.reshape(np.std(dataWait2[:,1::], axis=1), (numRates,1))), axis=1)
utilAvg2 = np.concatenate((1.0/utilVals2[:,0:1], np.reshape(np.mean(utilVals2[:,1::], axis=1), (numRates,1))), axis=1) 
utilStd2 = np.concatenate((1.0/utilVals2[:,0:1], np.reshape(np.std(utilVals2[:,1::], axis=1), (numRates,1))), axis=1) 
rates2 = (1.0/dataWait2[:,0:1])*4
       
dataWait3, utilVals3 = collate_mc_data(dataDir + "4bus-old/")
waitAvg3 = np.concatenate((1.0/dataWait3[:,0:1], np.reshape(np.mean(dataWait3[:,1::], axis=1), (numRates,1))), axis=1)
waitStd3 = np.concatenate((1.0/dataWait3[:,0:1], np.reshape(np.std(dataWait3[:,1::], axis=1), (numRates,1))), axis=1)
utilAvg3 = np.concatenate((1.0/utilVals3[:,0:1], np.reshape(np.mean(utilVals3[:,1::], axis=1), (numRates,1))), axis=1) 
utilStd3 = np.concatenate((1.0/utilVals3[:,0:1], np.reshape(np.std(utilVals3[:,1::], axis=1), (numRates,1))), axis=1) 
rates3 = (1.0/dataWait3[:,0:1])*4
"""
dataWait4, utilVals4 = collate_mc_data(dataDir + "3bus/")
waitAvg4 = np.concatenate((1.0/dataWait4[:,0:1], np.reshape(np.mean(dataWait4[:,1::], axis=1), (numRates,1))), axis=1)
waitStd4 = np.concatenate((1.0/dataWait4[:,0:1], np.reshape(np.std(dataWait4[:,1::], axis=1), (numRates,1))), axis=1)
utilAvg4 = np.concatenate((1.0/utilVals4[:,0:1], np.reshape(np.mean(utilVals4[:,1::], axis=1), (numRates,1))), axis=1) 
utilStd4 = np.concatenate((1.0/utilVals4[:,0:1], np.reshape(np.std(utilVals4[:,1::], axis=1), (numRates,1))), axis=1) 
rates4 = (1.0/dataWait4[:,0:1])*3
"""
```

```python
import scipy as sp
from scipy.interpolate import *

def getInterpF(utilVec, waitVec, windowLen=100):
    points = zip(list(utilVec), list(waitVec))
    points = sorted(points, key=lambda point: point[0])
    x1, y1 = zip(*points)
    returnF = interp1d(x1, y1, kind="linear")
    window = np.linspace(min(x1), max(x1), windowLen)
    return(returnF(window))

x1 = np.linspace(min(utilAvg2[:,1]), max(utilAvg2[:,1]), numRates)
y1 = np.reshape(getInterpF(utilAvg2[:,1], waitAvg2[:,1], windowLen=numRates), (numRates,1))
y2 = np.reshape(getInterpF(utilAvg2[:,1], (waitAvg2[:,1] + waitStd2[:,1]), windowLen=numRates), (numRates,1))
y3 = np.reshape(getInterpF(utilAvg2[:,1], (waitAvg2[:,1] - waitStd2[:,1]), windowLen=numRates), (numRates,1))

x2 = np.linspace(min(utilAvg3[:,1]), max(utilAvg3[:,1]), numRates)
y4 = np.reshape(getInterpF(utilAvg3[:,1], waitAvg3[:,1], windowLen=numRates), (numRates,1))
y5 = np.reshape(getInterpF(utilAvg3[:,1], (waitAvg3[:,1] + waitStd3[:,1]), windowLen=numRates), (numRates,1))
y6 = np.reshape(getInterpF(utilAvg3[:,1], (waitAvg3[:,1] - waitStd3[:,1]), windowLen=numRates), (numRates,1))


x3 = np.linspace(min(utilAvg4[:,1]), max(utilAvg4[:,1]), numRates)
y7 = np.reshape(getInterpF(utilAvg4[:,1], waitAvg4[:,1], windowLen=numRates), (numRates,1))
y8 = np.reshape(getInterpF(utilAvg4[:,1], (waitAvg4[:,1] + waitStd4[:,1]), windowLen=numRates), (numRates,1))
y9 = np.reshape(getInterpF(utilAvg4[:,1], (waitAvg4[:,1] - waitStd4[:,1]), windowLen=numRates), (numRates,1))
```
