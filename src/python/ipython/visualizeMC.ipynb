```python
import os
import numpy as np
import matplotlib.pyplot as plt
dataDir = "/home/chase/projects/net-queue/data/"
files = os.listdir(dataDir)
types = ["averageWait", "utilization"]


def collate_mc_data(dataPath, numBusses, simRates = 80):
    files = os.listdir(dataPath)
    waitSum = np.zeros((simRates, 2))
    utilSum = np.zeros((simRates, numBusses + 1))

    counterWait = 0
    counterUtil = 0
    for fileName in files:
        if fileName[0] == "a":
            datanext = np.loadtxt(dataPath + fileName, delimiter=",")
            waitSum += datanext
            counterWait += 1
        if fileName[0] == "u":
            datanextu = np.loadtxt(dataPath + fileName, delimiter=",")
            utilSum += datanextu
            counterUtil += 1
    
    waitSmooth = waitSum/float(counterWait)
    utilSmooth = utilSum/float(counterUtil)
    
    return(waitSmooth, utilSmooth)

#2 bus
dataSmooth2, utilSmooth2 = collate_mc_data(dataDir + "2bus/", 2)
          
#3 bus        
dataSmooth3, utilSmooth3 = collate_mc_data(dataDir + "3bus/", 3)
```

```python
def averageUtilization(utilizationArray):
    averageUtil = np.zeros((utilizationArray.shape[0], 1))
    numbusses = utilizationArray.shape[1]-1
    for i in range(numbusses):
        averageUtil = averageUtil + utilizationArray[:,i+1:i+2]
    averageUtil = averageUtil/float(numbusses)
    return(averageUtil)
        

averageUtil2 = averageUtilization(utilSmooth2)
averageUtil3 = averageUtilization(utilSmooth3)
```

```python
import scipy as sp
from scipy.interpolate import *

def getInterpF(utilVec, waitVec, windowLen=100):
    points = zip(list(utilVec), list(waitVec))
    points = sorted(points, key=lambda point: point[0])
    x1, y1 = zip(*points)
    returnF = interp1d(x1, y1, kind="linear")
    window = np.linspace(min(x1), max(x1), windowLen)
    return(returnF(window))

plt.plot(getInterpF(averageUtil2[:,0], dataSmooth2[:,1], windowLen=100), 'b-')
plt.plot(getInterpF(averageUtil3[:,0], dataSmooth3[:,1], windowLen=100), 'r-')
plt.xlabel("Average % blockface utilization")
plt.ylabel("Average wait w.r.t exogenous arrival rate")
plt.title("Average network wait time as function of blockface utilization")
plt.show()
```
